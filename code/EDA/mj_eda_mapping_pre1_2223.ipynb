```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime 
import os
from PIL import Image
import cv2
```

### 데이터 불러오기

```python
data = pd.read_csv("/home/ubuntu/drive/preprocessing/merged_growth_real_final_filtering2.csv", encoding="UTF-8")
data
```

```python
cdata = data.copy()
```

```python
cdata = cdata[cdata['is_filtering'] == 0]
cdata
```

```python
#cdata에서 필요한 데이터만 남기기
columns_to_keep = ['check_datetime', 'farm_code', 'area_code', 'input_image_name', 'input_image_path', 'before_img_name', 'img_id', 'count_masks', 'mask_num', 'mask_id','output_image_path', 'pred_growth']
clean_data = cdata[columns_to_keep]
clean_data
```

```python
#농가별로 데이터 구분하기
F0016_data = clean_data[clean_data['farm_code'] == 'F0016']
F0017_data = clean_data[clean_data['farm_code'] == 'F0017']
```

```python
#작기기간 정의 후, 농가별 - 작기별로 데이터 나누기 
# 작기 기간 설정

def period(start_date, end_date):
    start_date = datetime.strptime(start_date,'%Y-%m-%d %H:%M:%S')
    end_date = datetime.strptime(end_date, '%Y-%m-%d %H:%M:%S')
    return (start_date, end_date)

period_21_22 = period('2021-09-10 07:59:03', '2022-06-09 12:30:00')
period_22_23 = period('2022-08-25 16:44:07', '2023-06-23 16:36:06')
period_23_24 = period('2023-08-29 08:40:22', '2024-03-26 12:40:00')

# 농가별로 작기별 데이터를 만들어서 재저장
F0016_data['check_datetime'] = pd.to_datetime(F0016_data['check_datetime'])
F0016_period_21_22 = F0016_data[(F0016_data['check_datetime'] >= period_21_22[0]) & (F0016_data['check_datetime'] <= period_21_22[1])]
F0016_period_22_23 = F0016_data[(F0016_data['check_datetime'] >= period_22_23[0]) & (F0016_data['check_datetime'] <= period_22_23[1])]
F0016_period_23_24 = F0016_data[(F0016_data['check_datetime'] >= period_23_24[0]) & (F0016_data['check_datetime'] <= period_23_24[1])]

F0017_data['check_datetime'] = pd.to_datetime(F0017_data['check_datetime'])
F0017_period_21_22 = F0017_data[(F0017_data['check_datetime'] >= period_21_22[0]) & (F0017_data['check_datetime'] <= period_21_22[1])]
F0017_period_22_23 = F0017_data[(F0017_data['check_datetime'] >= period_22_23[0]) & (F0017_data['check_datetime'] <= period_22_23[1])]
F0017_period_23_24 = F0017_data[(F0017_data['check_datetime'] >= period_23_24[0]) & (F0017_data['check_datetime'] <= period_23_24[1])]
```

#### F0016 22-23 11월

```python
# 11월 데이터만 선택하여 저장
F0016_22_23_nov = F0016_period_22_23[(F0016_period_22_23['check_datetime'].dt.month == 11) & (F0016_period_22_23['check_datetime'].dt.year == 2022)]
F0016_22_23_nov

```

```python
F00162223nov = F0016_22_23_nov["mask_id"].value_counts().reset_index()
F00162223nov.columns = ['mask_id', 'count']  # 컬럼명 변경
F00162223nov_filtered = F00162223nov[F00162223nov["count"] > 2]
```

```python


def save_figure_as_image(fig, output_path):
    # Figure 객체를 이미지로 변환하여 저장
    fig.savefig(output_path, bbox_inches='tight', pad_inches=0)
    plt.close(fig)

# F00162223feb_filtered에 담긴 mask_id를 사용하여 데이터 필터링
filtered_data = F0016_period_22_23[F0016_period_22_23['mask_id'].isin(F00162223nov_filtered['mask_id'])]

# mask_id별로 최신 날짜순으로 정렬하고 필요한 정보 저장
mask_info_list = []
for mask_id in F00162223nov_filtered['mask_id']:
    mask_data = filtered_data[filtered_data['mask_id'] == mask_id]
    sorted_mask_data = mask_data.sort_values(by='check_datetime', ascending=True)
    mask_info_list.append(sorted_mask_data)

# 각 mask_id별로 폴더를 생성하고 이미지 저장
for mask_info, mask_id in zip(mask_info_list, F00162223nov_filtered['mask_id']):
    # 새로운 Figure 생성
    fig, axes = plt.subplots(1, len(mask_info), figsize=(28, 6))
    fig.suptitle(f'mask_id: {mask_id}', fontsize=15)

    for i, (_, row) in enumerate(mask_info.iterrows()):
        img_path = "/home/ubuntu/drive/dataset/SAM_result/F0016/C101/" + row['input_image_name'] + "/masks/" + row["output_image_path"]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # 이미지와 관련 정보 시각화
        check_datetime_formatted_year = row["check_datetime"].strftime("%Y")
        check_datetime_formatted_day = row["check_datetime"].strftime("%m-%d")
        check_datetime_formatted = row["check_datetime"].strftime("%H:%M")
        pred_growth = row["pred_growth"]
        title = f'{check_datetime_formatted_year}\n{check_datetime_formatted_day}\n{check_datetime_formatted}\n{pred_growth}'
        axes[i].imshow(img)
        axes[i].set_title(title, fontsize=10)
        axes[i].axis('off')

    # 이미지 파일로 저장
    check_datetime_formatted_day = mask_info.iloc[0]["check_datetime"].strftime("%m-%d")  # 루프 첫 번째 행의 날짜 사용
    output_folder = "/home/ubuntu/drive/EDA/mj_mapping_pre1_2223/22-11/" + mask_id + "/"
    os.makedirs(output_folder, exist_ok=True)  # mask_id 별 폴더 생성
    output_image_path = output_folder + f"{mask_id}_{check_datetime_formatted_day}.png"
    save_figure_as_image(fig, output_image_path)

    print(f"이미지가 성공적으로 저장되었습니다: {output_image_path}")
```
